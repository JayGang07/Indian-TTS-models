```python
from transformers import pipeline
import scipy.io.wavfile
from IPython.display import Audio, display

# 1. Initialize the Text-to-Speech pipeline using Meta MMS English
# (This will automatically download the lightweight ~145MB model)
pipe = pipeline("text-to-speech", model="facebook/mms-tts-hin")

# 2. Enter the text prompt you want the AI to speak
text_prompt = "नमस्कार, मेरा नाम जय है।"

# 3. Generate the speech audio
output = pipe(text_prompt)

# 4. Play the audio directly inside your Jupyter/Colab notebook cell
display(Audio(output["audio"], rate=output["sampling_rate"]))

# 5. Save the audio to a local file
scipy.io.wavfile.write("mms_hindi_test.wav", rate=output["sampling_rate"], data=output["audio"])
print("Hindi audio successfully generated and saved as 'mms_hindi_test.wav'!")
```
