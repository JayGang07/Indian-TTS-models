```python
import os
import json
import whisper
import jiwer
import requests
import time
import numpy as np

# --- Configuration ---
API_KEY = "vach_1ytE2CY5X2EPdZgZlDBbi39qeHPJzxX3MWAYEyvf82f5RT6BFzQTTV2PTv3VDgvT3fs2nA5kMEpprB9ea3kjYJO7T13GcGJf_d45996d1e60a19972a97486d5c5955d3"
GNANI_ASR_ENDPOINT = "https://api.vachana.ai/api/v1/asr/rest"

print("Loading Whisper medium model...")
whisper_model = whisper.load_model("medium")

def batch_evaluate_hindi():
    print("\n" + "="*40)
    print("🚀 Starting Batch Evaluation: HINDI (Whisper)")
    print("="*40)

    # Load ground truth map
    with open('/content/hindi_evaluation_set.json', 'r', encoding='utf-8') as f:
        hin_data = json.load(f)

    # Map filenames to their reference text
    reference_map = {}
    for key, item in hin_data.items():
        # Reconstruct the exact filename we saved earlier
        filename = f"HIN_{item['id']}_{key}.wav"
        reference_map[filename] = item["text"]

    audio_dir = "/content/output_audio/hindi"
    wers, cers = [], []

    for filename in os.listdir(audio_dir):
        if not filename.endswith(".wav"): continue

        audio_path = os.path.join(audio_dir, filename)
        reference_text = reference_map.get(filename)

        if not reference_text:
            continue

        # Transcribe
        result = whisper_model.transcribe(audio_path, language="hi")
        hypothesis_text = result["text"].strip()

        # Calculate
        wer = jiwer.wer(reference_text, hypothesis_text)
        cer = jiwer.cer(reference_text, hypothesis_text)
        wers.append(wer)
        cers.append(cer)

        print(f"File: {filename} | WER: {wer:.2f} | CER: {cer:.2f}")

    if wers:
        print(f"\n📊 HINDI AVERAGE -> WER: {np.mean(wers):.4f} | CER: {np.mean(cers):.4f}")

def batch_evaluate_bengali():
    print("\n" + "="*40)
    print("🚀 Starting Batch Evaluation: BENGALI (Gnani ASR)")
    print("="*40)

    # Load ground truth map
    with open('/content/bengali_evaluation_set.json', 'r', encoding='utf-8') as f:
        ben_data = json.load(f)

    reference_map = {f"BEN_{item['id']}.wav": item["bengali_sentence"] for item in ben_data}

    audio_dir = "/content/output_audio/bengali"
    headers = {"X-API-Key-ID": API_KEY}
    wers, cers = [], []

    for filename in sorted(os.listdir(audio_dir)):
        if not filename.endswith(".wav"): continue

        audio_path = os.path.join(audio_dir, filename)
        reference_text = reference_map.get(filename)

        if not reference_text:
            continue

        try:
            with open(audio_path, 'rb') as audio_file:
                response = requests.post(
                    GNANI_ASR_ENDPOINT,
                    headers=headers,
                    files={'file': audio_file},
                    data={'language': 'bn-IN'}
                )
                response.raise_for_status()

                hypothesis_text = response.json().get('transcript', '').strip()

                wer = jiwer.wer(reference_text, hypothesis_text)
                cer = jiwer.cer(reference_text, hypothesis_text)
                wers.append(wer)
                cers.append(cer)

                print(f"File: {filename} | WER: {wer:.2f} | CER: {cer:.2f}")

                # Tiny buffer to prevent hitting Gnani rate limits
                time.sleep(0.3)

        except Exception as e:
            print(f"❌ Failed on {filename}: {e}")

    if wers:
        print(f"\n📊 BENGALI AVERAGE -> WER: {np.mean(wers):.4f} | CER: {np.mean(cers):.4f}")

if __name__ == "__main__":
    batch_evaluate_hindi()
    batch_evaluate_bengali()
```

```python
import io
import os
import json
import csv
import time
import requests
import editdistance
from pathlib import Path
from pydub import AudioSegment

# --- Configuration based on Reference Script ---
API_KEY = "vach_1ytE2CY5X2EPdZgZlDBbi39qeHPJzxX3MWAYEyvf82f5RT6BFzQTTV2PTv3VDgvT3fs2nA5kMEpprB9ea3kjYJO7T13GcGJf_d45996d1e60a19972a97486d5c5955d3"
API_URL = "https://api.vachana.ai/stt/v3"
LANGUAGE_CODE = "bn-IN"
FORMAT = "transcribe"
REQUEST_DELAY_SEC = 1.2

AUDIO_DIR = Path("/content/output_audio/bengali")
JSON_PATH = Path("/content/bengali_evaluation_set.json")
OUTPUT_CSV = Path("/content/bengali_evaluation_results.csv")

# --- Helper Functions ---
def normalize(text: str) -> str:
    return " ".join(text.strip().lower().split())

def compute_wer(ref: str, hyp: str) -> float:
    r, h = normalize(ref).split(), normalize(hyp).split()
    return editdistance.eval(r, h) / len(r) if r else 0.0

def compute_cer(ref: str, hyp: str) -> float:
    r = list(normalize(ref).replace(" ", ""))
    h = list(normalize(hyp).replace(" ", ""))
    return editdistance.eval(r, h) / len(r) if r else 0.0

def transcribe(audio_path: Path, api_key: str) -> str:
    headers = {"X-API-Key-ID": api_key}
    data = {"language_code": LANGUAGE_CODE, "format": FORMAT}

    # Convert to 16kHz mono WAV in memory (Crucial for Prisma v2.5)
    audio = AudioSegment.from_file(str(audio_path))
    audio = audio.set_frame_rate(16000).set_channels(1).set_sample_width(2)
    buf = io.BytesIO()
    audio.export(buf, format="wav")
    buf.seek(0)

    files = {"audio_file": (audio_path.name, buf, "audio/wav")}

    try:
        resp = requests.post(API_URL, headers=headers, data=data, files=files, timeout=60)
        resp.raise_for_status()
        payload = resp.json()

        if not payload.get("success", True):
            print(f"  [API ERROR] {payload}")
            return None

        return payload.get("transcript", "")

    except Exception as e:
        print(f"  [HTTP/REQUEST ERROR] {e}")
        return None

# --- Main Evaluation Loop ---
def run_evaluation():
    print(f"\n{'='*60}")
    print(f"  🚀 Starting Gnani ASR (v3) Evaluation for Bengali")
    print(f"{'='*60}")

    # Load references
    with open(JSON_PATH, 'r', encoding='utf-8') as f:
        ben_data = json.load(f)

    reference_map = {f"BEN_{item['id']}.wav": item["bengali_sentence"] for item in ben_data}

    # Prepare CSV
    csv_exists = OUTPUT_CSV.exists()
    csv_file = open(OUTPUT_CSV, "a", newline="", encoding="utf-8")
    writer = csv.DictWriter(csv_file, fieldnames=["filename", "reference", "hypothesis", "wer", "cer", "status"])
    if not csv_exists:
        writer.writeheader()

    wer_scores, cer_scores = [], []
    error_count = 0

    wav_files = sorted([f for f in AUDIO_DIR.iterdir() if f.suffix == '.wav'])
    n = len(wav_files)

    for i, wav_path in enumerate(wav_files):
        filename = wav_path.name
        reference = reference_map.get(filename)

        if not reference:
            print(f"  [{i+1:>3}/{n}] SKIPPED (No reference found for {filename})")
            continue

        print(f"  [{i+1:>3}/{n}] {filename} ... ", end="", flush=True)

        hypothesis = transcribe(wav_path, API_KEY)

        if hypothesis is None:
            print("FAILED ❌")
            error_count += 1
            writer.writerow({
                "filename": filename, "reference": reference, "hypothesis": "",
                "wer": "", "cer": "", "status": "error"
            })
        else:
            wer = compute_wer(reference, hypothesis)
            cer = compute_cer(reference, hypothesis)
            wer_scores.append(wer)
            cer_scores.append(cer)

            print(f"WER={wer:.3f} | CER={cer:.3f} ✅")
            writer.writerow({
                "filename": filename, "reference": reference, "hypothesis": hypothesis,
                "wer": round(wer, 4), "cer": round(cer, 4), "status": "ok"
            })

        csv_file.flush()
        time.sleep(REQUEST_DELAY_SEC)

    csv_file.close()

    # Final Summary
    success = len(wer_scores)
    avg_wer = sum(wer_scores) / success if success else 0.0
    avg_cer = sum(cer_scores) / success if success else 0.0

    print(f"\n{'='*60}")
    print(f"  📊 Final Bengali Report")
    print(f"{'='*60}")
    print(f"  Total Processed : {success + error_count}")
    print(f"  Successful      : {success}")
    print(f"  Errors          : {error_count}")
    if success > 0:
        print(f"  Average WER     : {avg_wer:.4f}")
        print(f"  Average CER     : {avg_cer:.4f}")
    print(f"  Results saved to: {OUTPUT_CSV}")

if __name__ == "__main__":
    run_evaluation()
```
