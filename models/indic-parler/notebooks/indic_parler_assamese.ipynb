```python
import os
import json
import torch
import soundfile as sf
from transformers import AutoTokenizer
from parler_tts import ParlerTTSForConditionalGeneration

# ---------------------------------------------------------
# 1. Load the Assamese Dataset from the Colab Side Panel
# ---------------------------------------------------------
file_path = "assamese_evaluation_set.json"

if not os.path.exists(file_path):
    raise FileNotFoundError(f"Please upload '{file_path}' to the Colab side panel before running this cell.")

print(f"Loading dataset from {file_path}...")
with open(file_path, "r", encoding="utf-8") as f:
    assamese_dataset = json.load(f)

# ---------------------------------------------------------
# 2. Initialize Device and Model
# ---------------------------------------------------------
device = "cuda:0" if torch.cuda.is_available() else "cpu"
model_id = "ai4bharat/indic-parler-tts"

print(f"Loading {model_id} on {device}...")
model = ParlerTTSForConditionalGeneration.from_pretrained(model_id).to(device)
tokenizer = AutoTokenizer.from_pretrained(model_id)

# ---------------------------------------------------------
# 3. Define the Voice Description
# ---------------------------------------------------------
base_description = "A native Assamese speaker with a clear voice, delivering an expressive speech at a normal pace in a quiet environment."

# ---------------------------------------------------------
# 4. Processing Loop and Audio Generation
# ---------------------------------------------------------
output_dir = "assamese_evaluation_outputs"
os.makedirs(output_dir, exist_ok=True)

print(f"Starting TTS generation for {len(assamese_dataset)} samples...\n")

for item in assamese_dataset:
    idx = item["id"]
    text = item["assamese_sentence"]
    category = item["benchmark_category"]
    targets = item["phonetic_targets"]

    print(f"--- Processing ID {idx} ---")
    print(f"Category: {category}")
    print(f"Targets: {targets}")
    print(f"Text: {text}")

    # Adjust description dynamically based on the benchmark category
    current_description = base_description
    if "Anger" in category:
        current_description = "An angry Assamese speaker delivering a fast-paced, high-energy speech."
    elif "Sadness" in category:
        current_description = "A sad Assamese speaker delivering a slow-paced, low-pitched speech."
    elif "Joy" in category:
        current_description = "A joyful Assamese speaker delivering a fast-paced, highly expressive speech."

    # Tokenize the description and the text prompt
    input_ids = tokenizer(current_description, return_tensors="pt").input_ids.to(device)
    prompt_input_ids = tokenizer(text, return_tensors="pt").input_ids.to(device)

    # Generate the audio tensor
    with torch.no_grad():
        generation = model.generate(
            input_ids=input_ids,
            prompt_input_ids=prompt_input_ids
        )

    # Convert tensor to numpy array for saving
    audio_arr = generation.cpu().numpy().squeeze()
    sample_rate = model.config.sampling_rate

    # Save the file
    output_filename = os.path.join(output_dir, f"assamese_benchmark_{idx:02d}.wav")
    sf.write(output_filename, audio_arr, sample_rate)

    print(f"Saved: {output_filename}\n")

print("All Assamese benchmark tests have been synthesized successfully.")
```

```python
import os
import json
import torch
import soundfile as sf
import jiwer
from transformers import AutoTokenizer, pipeline
from parler_tts import ParlerTTSForConditionalGeneration

# ---------------------------------------------------------
# 1. Load the Assamese Dataset from the Colab Side Panel
# ---------------------------------------------------------
file_path = "assamese_evaluation_set.json"

if not os.path.exists(file_path):
    raise FileNotFoundError(f"Please upload '{file_path}' to the Colab side panel before running this cell.")

print(f"Loading dataset from {file_path}...")
with open(file_path, "r", encoding="utf-8") as f:
    assamese_dataset = json.load(f)

# ---------------------------------------------------------
# 2. Initialize Device, TTS Model, and ASR Pipeline
# ---------------------------------------------------------
device = "cuda:0" if torch.cuda.is_available() else "cpu"

# TTS Setup
tts_model_id = "ai4bharat/indic-parler-tts"
print(f"Loading TTS model {tts_model_id} on {device}...")
tts_model = ParlerTTSForConditionalGeneration.from_pretrained(tts_model_id).to(device)
tts_tokenizer = AutoTokenizer.from_pretrained(tts_model_id)

# ASR Setup (Whisper small is generally a good balance of speed/accuracy for Colab)
# ASR Setup (Updated to handle long-form audio > 30s)
asr_model_id = "openai/whisper-small"
print(f"Loading ASR model {asr_model_id} on {device}...")
asr_pipe = pipeline(
    "automatic-speech-recognition",
    model=asr_model_id,
    device=device,
    generate_kwargs={
        "language": "assamese",
        "return_timestamps": True  # Fixes the > 30-second error
    }
)
# ---------------------------------------------------------
# 3. Define Evaluation Preprocessing
# ---------------------------------------------------------
# To ensure punctuation and casing don't artificially inflate error rates
text_transform = jiwer.Compose([
    jiwer.ToLowerCase(),
    jiwer.RemovePunctuation(),
    jiwer.RemoveMultipleSpaces(),
    jiwer.Strip(),
])

base_description = "A native Assamese speaker with a clear voice, delivering an expressive speech at a normal pace in a quiet environment."

# ---------------------------------------------------------
# 4. Processing Loop (Generation + ASR + Metrics)
# ---------------------------------------------------------
output_dir = "assamese_evaluation_outputs"
os.makedirs(output_dir, exist_ok=True)
results_log = []

print(f"\nStarting TTS generation and evaluation for {len(assamese_dataset)} samples...\n")

for item in assamese_dataset:
    idx = item["id"]
    reference_text = item["assamese_sentence"]
    category = item["benchmark_category"]

    print(f"--- Processing ID {idx} ---")

    # Adjust description dynamically
    current_description = base_description
    if "Anger" in category:
        current_description = "An angry Assamese speaker delivering a fast-paced, high-energy speech."
    elif "Sadness" in category:
        current_description = "A sad Assamese speaker delivering a slow-paced, low-pitched speech."
    elif "Joy" in category:
        current_description = "A joyful Assamese speaker delivering a fast-paced, highly expressive speech."

    # TTS Generation
    input_ids = tts_tokenizer(current_description, return_tensors="pt").input_ids.to(device)
    prompt_input_ids = tts_tokenizer(reference_text, return_tensors="pt").input_ids.to(device)

    with torch.no_grad():
        generation = tts_model.generate(input_ids=input_ids, prompt_input_ids=prompt_input_ids)

    audio_arr = generation.cpu().numpy().squeeze()
    sample_rate = tts_model.config.sampling_rate

    # Save the audio file
    output_filename = os.path.join(output_dir, f"assamese_benchmark_{idx:02d}.wav")
    sf.write(output_filename, audio_arr, sample_rate)

    # ASR Transcription
    # ASR Transcription (Updated)
    transcription_result = asr_pipe(
        {"array": audio_arr, "sampling_rate": sample_rate},
        generate_kwargs={"language": "assamese", "return_timestamps": True}
    )
    hypothesis_text = transcription_result["text"]

    # Clean text for accurate metric calculation
    clean_reference = text_transform(reference_text)
    clean_hypothesis = text_transform(hypothesis_text)

    # Calculate Metrics
    # Fallback to avoid division by zero if reference is empty
    if len(clean_reference) > 0:
        wer_score = jiwer.wer(clean_reference, clean_hypothesis)
        cer_score = jiwer.cer(clean_reference, clean_hypothesis)
    else:
        wer_score, cer_score = 0.0, 0.0

    print(f"Original Text: {reference_text}")
    print(f"Transcribed:   {hypothesis_text}")
    print(f"WER: {wer_score:.4f} | CER: {cer_score:.4f}\n")

    # Log results
    results_log.append({
        "id": idx,
        "benchmark_category": category,
        "reference_text": reference_text,
        "transcribed_text": hypothesis_text,
        "wer": wer_score,
        "cer": cer_score
    })

# ---------------------------------------------------------
# 5. Save the Evaluation Log
# ---------------------------------------------------------
log_path = os.path.join(output_dir, "evaluation_metrics_log.json")
with open(log_path, "w", encoding="utf-8") as f:
    json.dump(results_log, f, ensure_ascii=False, indent=4)

print(f"Evaluation complete. Audio files and metrics log saved to '{output_dir}'.")
```
