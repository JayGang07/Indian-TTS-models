```python
import os
import json
import torch
import scipy.io.wavfile
import whisper
import jiwer
import re
import shutil
from google.colab import files

# Parler-TTS specific imports
from parler_tts import ParlerTTSForConditionalGeneration
from transformers import AutoTokenizer

def normalize_bengali_text(text):
    """Removes standard and Bengali-specific punctuation for fairer WER/CER calculation."""
    text = re.sub(r'[।,\-\?!\"\'\.]', '', text)
    return re.sub(r'\s+', ' ', text).strip()

# ==========================================
# 1. Initialize Models
# ==========================================
print("Loading AI4Bharat Indic Parler-TTS...")
model_id = "ai4bharat/indic-parler-tts"
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load the custom Parler-TTS model and tokenizer
tts_model = ParlerTTSForConditionalGeneration.from_pretrained(model_id).to(device)
tokenizer = AutoTokenizer.from_pretrained(model_id)

print("Loading Whisper 'medium' ASR... (This will take a moment)")
asr_model = whisper.load_model("medium")

# ==========================================
# 2. Load Dataset & Prepare Environment
# ==========================================
with open("bengali_evaluation_set.json", "r", encoding="utf-8") as f:
    eval_dataset = json.load(f)

output_dir = "indic_parler_bengali_audio"
os.makedirs(output_dir, exist_ok=True)

# Define the Voice Persona (Parler-TTS requires this to shape the audio)
voice_description = "A female speaker with a clear, neutral tone and no emotional emphasis. The recording is very high quality with no background noise."
input_ids = tokenizer(voice_description, return_tensors="pt").input_ids.to(device)

print("\n" + "=" * 80)
print("STARTING PER-SENTENCE EVALUATION: INDIC PARLER-TTS")
print("=" * 80)

# ==========================================
# 3. Generate, Transcribe, and Evaluate
# ==========================================
# ==========================================
# 3. Generate, Transcribe, and Evaluate
# ==========================================
# Tokenize description with its mask
description_inputs = tokenizer(voice_description, return_tensors="pt").to(device)
input_ids = description_inputs.input_ids
attention_mask = description_inputs.attention_mask

print("\n" + "=" * 80)
print("STARTING PER-SENTENCE EVALUATION: INDIC PARLER-TTS")
print("=" * 80)

for item in eval_dataset:
    item_id = item["id"]
    raw_reference = item["bengali_sentence"]
    category = item["benchmark_category"]

    # --- Generate Audio ---
    # Tokenize the Bengali text along with its attention mask
    prompt_inputs = tokenizer(raw_reference, return_tensors="pt").to(device)

    with torch.no_grad():
        generation = tts_model.generate(
            input_ids=input_ids,
            attention_mask=attention_mask,
            prompt_input_ids=prompt_inputs.input_ids,
            prompt_attention_mask=prompt_inputs.attention_mask
        )

    # Extract waveform and save to file
    audio_arr = generation.cpu().numpy().squeeze()
    sample_rate = tts_model.config.sampling_rate
    audio_path = os.path.join(output_dir, f"eval_{item_id:02d}.wav")

    scipy.io.wavfile.write(audio_path, rate=sample_rate, data=audio_arr)

    # --- Transcribe Audio ---
    result = asr_model.transcribe(audio_path, language="bn")
    raw_hypothesis = result["text"]

    # --- Normalize Text ---
    norm_ref = normalize_bengali_text(raw_reference)
    norm_hyp = normalize_bengali_text(raw_hypothesis)

    # --- Calculate Metrics ---
    try:
        wer = jiwer.wer(norm_ref, norm_hyp)
        cer = jiwer.cer(norm_ref, norm_hyp)
    except ValueError:
        wer, cer = 1.0, 1.0

    # --- Output Results ---
    print(f"ID {item_id} | {category}")
    print(f"  Reference  : {norm_ref}")
    print(f"  Prediction : {norm_hyp}")
    print(f"  --> WER: {wer:.4f} ({wer * 100:.1f}%) | CER: {cer:.4f} ({cer * 100:.1f}%)\n")
    print("-" * 80)
# ==========================================
# 4. Zip and Download the Audio Folder
# ==========================================
zip_name = f"{output_dir}.zip"
print(f"\nZipping '{output_dir}' and downloading to local machine...")
shutil.make_archive(output_dir, 'zip', output_dir)
files.download(zip_name)
```
