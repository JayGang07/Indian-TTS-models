```python
import os
import json
import torch
import scipy.io.wavfile
import whisper
import jiwer
import re
from transformers import AutoModel, AutoTokenizer
from huggingface_hub import login

# ==========================================
# 1. Authenticate Explicitly
# ==========================================
# Replace with your actual Hugging Face token
login(token="your_hf_token_here")

def normalize_bengali_text(text):
    """Removes standard and Bengali-specific punctuation for fairer WER/CER calculation."""
    text = re.sub(r'[।,\-\?!\"\'\.]', '', text)
    return re.sub(r'\s+', ' ', text).strip()

# ==========================================
# 2. Initialize Models (Bypassing Pipeline)
# ==========================================
from transformers import AutoConfig, AutoModel, AutoTokenizer

# ==========================================
# 2. Initialize Models (Bypassing Pipeline)
# ==========================================
print("\nLoading AI4Bharat VITS-Rasa-13 TTS...")
model_id = "ai4bharat/vits_rasa_13"

# Step A: Fetch the raw configuration
config = AutoConfig.from_pretrained(model_id, trust_remote_code=True)

# Step B: Manually inject the missing padding token ID into the config object
config.pad_token_id = 0

# Step C: Initialize the model using the patched config
tts_model = AutoModel.from_pretrained(model_id, config=config, trust_remote_code=True)
tts_tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)

# Move TTS model to GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
tts_model = tts_model.to(device)

print("Loading Whisper 'medium' ASR... (This will take a moment)")
asr_model = whisper.load_model("medium")

# ==========================================
# 3. Load Dataset & Prepare Directory
# ==========================================
with open("bengali_evaluation_set.json", "r", encoding="utf-8") as f:
    eval_dataset = json.load(f)

output_dir = "vits_rasa_bengali_audio"
os.makedirs(output_dir, exist_ok=True)

print("\n" + "=" * 80)
print("STARTING PER-SENTENCE EVALUATION: AI4BHARAT VITS-RASA-13")
print("=" * 80)

# IMPORTANT NOTE ON SPEAKER IDs:
# AI4Bharat uses specific integer IDs for each language/gender combination.
# E.g., Punjabi Male is 16. Bengali usually falls in the 2, 3, or 4 range.
# You may need to tweak this integer to get the exact Bengali voice profile.
speaker_id = 2
style_id = 0 # 0 is usually the default neutral/Alexa style

# ==========================================
# 4. Generate, Transcribe, and Evaluate
# ==========================================
for item in eval_dataset:
    item_id = item["id"]
    raw_reference = item["bengali_sentence"]
    category = item["benchmark_category"]

    # --- Generate Audio ---
    inputs = tts_tokenizer(text=raw_reference, return_tensors="pt").to(device)

    # Run inference without tracking gradients (saves memory)
    with torch.no_grad():
        outputs = tts_model(inputs['input_ids'], speaker_id=speaker_id, emotion_id=style_id)

    # Extract the waveform tensor, squeeze it, and move it to CPU as a numpy array
    waveform = outputs.waveform.squeeze().cpu().numpy()

    audio_path = os.path.join(output_dir, f"eval_{item_id:02d}.wav")
    scipy.io.wavfile.write(audio_path, rate=tts_model.config.sampling_rate, data=waveform)

    # --- Transcribe Audio ---
    result = asr_model.transcribe(audio_path, language="bn")
    raw_hypothesis = result["text"]

    # --- Normalize Text ---
    norm_ref = normalize_bengali_text(raw_reference)
    norm_hyp = normalize_bengali_text(raw_hypothesis)

    # --- Calculate Metrics ---
    try:
        wer = jiwer.wer(norm_ref, norm_hyp)
        cer = jiwer.cer(norm_ref, norm_hyp)
    except ValueError:
        # Failsafe if Whisper outputs an empty string
        wer, cer = 1.0, 1.0

    # --- Output Results ---
    print(f"ID {item_id} | {category}")
    print(f"  Reference  : {norm_ref}")
    print(f"  Prediction : {norm_hyp}")
    print(f"  --> WER: {wer:.4f} ({wer * 100:.1f}%) | CER: {cer:.4f} ({cer * 100:.1f}%)\n")
    print("-" * 80)
```
