```python
import os
import json
import torch
import soundfile as sf
import numpy as np
import whisper
import jiwer
import re
import shutil
from datasets import load_dataset
from transformers import AutoModel
from google.colab import files


def normalize_assamese_text(text):
    text = re.sub(r'[।,\-\?!\"\'\.]', '', text)
    return re.sub(r'\s+', ' ', text).strip()

# ==========================================
# 1. Fetch Reference Audio
# ==========================================
print("\nFetching reference audio...")
TARGET_SPEAKER = "S4258130900301480"
REF_AUDIO_PATH = "assamese_reference.wav"

if not os.path.exists(REF_AUDIO_PATH):
    dataset = load_dataset("ai4bharat/indicvoices_r", "Assamese", split="train", streaming=True)
    for item in dataset:
        if item["speaker_id"] == TARGET_SPEAKER:
            sf.write(REF_AUDIO_PATH, item["audio"]["array"], item["audio"]["sampling_rate"])
            REF_TEXT = item["normalized"]
            break
else:
    print(f"'{REF_AUDIO_PATH}' already exists. Skipping download.")
    REF_TEXT = "স্কুলত মইটো এতিয়া স্কুলত গৈ পেলাই ষাঠিটামানৰ ল ৰা ছোৱালীৰ ভাতেই বনাব লাগে"

print(f"  -> Reference Text: {REF_TEXT}")

# ==========================================
# 2. Load Models
# ==========================================
device = "cuda:0" if torch.cuda.is_available() else "cpu"
print(f"\nUsing device: {device}")

print("Loading AI4Bharat IndicF5 TTS...")
tts_model = AutoModel.from_pretrained(
    "ai4bharat/IndicF5",
    trust_remote_code=True,
    torch_dtype=torch.float32
)
tts_model = tts_model.to(device)

print("Loading Whisper 'medium' ASR...")
asr_model = whisper.load_model("medium").to(device)
print("Both models ready.")

# ==========================================
# 3. Load Dataset & Prepare Output Dir
# ==========================================
with open("assamese_evaluation_set.json", "r", encoding="utf-8") as f:
    eval_dataset = json.load(f)

output_dir = "indic_f5_assamese_audio"
os.makedirs(output_dir, exist_ok=True)

print("\n" + "=" * 80)
print("STARTING PER-SENTENCE EVALUATION: INDIC F5 (ASSAMESE)")
print("=" * 80)

# ==========================================
# 4. Generate, Transcribe, Evaluate
# ==========================================
for item in eval_dataset:
    item_id = item["id"]
    raw_reference = item["assamese_sentence"]
    category = item["benchmark_category"]

    try:
        with torch.no_grad():
            audio = tts_model(
                raw_reference,
                ref_audio_path=REF_AUDIO_PATH,
                ref_text=REF_TEXT
            )

        if audio.dtype == np.int16:
            audio = audio.astype(np.float32) / 32768.0

        audio_path = os.path.join(output_dir, f"eval_{item_id:02d}.wav")
        # soundfile replaces scipy.io.wavfile — no numpy version conflict
        sf.write(audio_path, np.array(audio, dtype=np.float32), 24000, subtype='FLOAT')

        result = asr_model.transcribe(audio_path, language="as")
        raw_hypothesis = result["text"]

        norm_ref = normalize_assamese_text(raw_reference)
        norm_hyp = normalize_assamese_text(raw_hypothesis)

        wer = jiwer.wer(norm_ref, norm_hyp) if len(norm_ref) > 0 else 1.0
        cer = jiwer.cer(norm_ref, norm_hyp) if len(norm_ref) > 0 else 1.0

        print(f"ID {item_id} | {category}")
        print(f"  Reference  : {norm_ref}")
        print(f"  Prediction : {norm_hyp}")
        print(f"  --> WER: {wer:.4f} ({wer*100:.1f}%) | CER: {cer:.4f} ({cer*100:.1f}%)\n")
        print("-" * 80)

    except Exception as e:
        print(f"Error processing ID {item_id}: {e}\n")
        print("-" * 80)

# ==========================================
# 5. Zip and Download
# ==========================================
zip_name = f"{output_dir}.zip"
print(f"\nZipping and downloading '{output_dir}'...")
shutil.make_archive(output_dir, 'zip', output_dir)
files.download(zip_name)
```
