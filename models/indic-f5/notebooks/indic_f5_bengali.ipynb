```python
import os
import json
import torch
import soundfile as sf
import scipy.io.wavfile
import numpy as np
import whisper
import jiwer
import re
import shutil
from datasets import load_dataset
from transformers import AutoModel
from google.colab import files
from huggingface_hub import login

# ==========================================
# 0. Authenticate with Hugging Face
# ==========================================
login(token="your_hf_token_here") # Replace with your actual token

def normalize_bengali_text(text):
    """Removes standard and Bengali-specific punctuation for fairer WER/CER calculation."""
    text = re.sub(r'[।,\-\?!\"\'\.]', '', text)
    return re.sub(r'\s+', ' ', text).strip()

# ==========================================
# 1. Fetch Reference from IndicVoices-R
# ==========================================
print("\nFetching reference audio and text from ai4bharat/indicvoices_r...")

# Stream the dataset to avoid downloading the entire Bengali split
dataset = load_dataset("ai4bharat/indicvoices_r", "Bengali", split="train", streaming=True)
ref_sample = next(iter(dataset))

# Extract the text and audio array
REF_TEXT = ref_sample["text"]
audio_array = ref_sample["audio"]["array"]
sampling_rate = ref_sample["audio"]["sampling_rate"]

# Save the audio locally so IndicF5 can use it
REF_AUDIO_PATH = "bengali_reference.wav"
sf.write(REF_AUDIO_PATH, audio_array, sampling_rate)

print(f"  -> Reference Text: {REF_TEXT}")
print(f"  -> Saved reference audio to '{REF_AUDIO_PATH}'")

# ==========================================
# 2. Initialize Models
# ==========================================
print("\nLoading AI4Bharat IndicF5 TTS...")
# Because we are on transformers==4.49.0, this will now load natively without crashing
tts_model = AutoModel.from_pretrained(
    "ai4bharat/IndicF5",
    trust_remote_code=True
)

device = "cuda" if torch.cuda.is_available() else "cpu"
tts_model = tts_model.to(device)

print("Loading Whisper 'medium' ASR... (This will take a moment)")
asr_model = whisper.load_model("medium")

# ==========================================
# 3. Load Dataset & Prepare Environment
# ==========================================
with open("bengali_evaluation_set.json", "r", encoding="utf-8") as f:
    eval_dataset = json.load(f)

output_dir = "indic_f5_bengali_audio"
os.makedirs(output_dir, exist_ok=True)

print("\n" + "=" * 80)
print("STARTING PER-SENTENCE EVALUATION: INDIC F5")
print("=" * 80)

# ==========================================
# 4. Generate, Transcribe, and Evaluate
# ==========================================
for item in eval_dataset:
    item_id = item["id"]
    raw_reference = item["bengali_sentence"]
    category = item["benchmark_category"]

    # --- Generate Audio ---
    with torch.no_grad():
        audio = tts_model(
            raw_reference,
            ref_audio_path=REF_AUDIO_PATH,
            ref_text=REF_TEXT
        )

    # Normalize the waveform array
    if audio.dtype == np.int16:
        audio = audio.astype(np.float32) / 32768.0

    audio_path = os.path.join(output_dir, f"eval_{item_id:02d}.wav")

    # Write the output at the standard 24kHz
    scipy.io.wavfile.write(audio_path, 24000, np.array(audio, dtype=np.float32))

    # --- Transcribe Audio ---
    result = asr_model.transcribe(audio_path, language="bn")
    raw_hypothesis = result["text"]

    # --- Normalize Text ---
    norm_ref = normalize_bengali_text(raw_reference)
    norm_hyp = normalize_bengali_text(raw_hypothesis)

    # --- Calculate Metrics ---
    try:
        wer = jiwer.wer(norm_ref, norm_hyp)
        cer = jiwer.cer(norm_ref, norm_hyp)
    except ValueError:
        wer, cer = 1.0, 1.0

    # --- Output Results ---
    print(f"ID {item_id} | {category}")
    print(f"  Reference  : {norm_ref}")
    print(f"  Prediction : {norm_hyp}")
    print(f"  --> WER: {wer:.4f} ({wer * 100:.1f}%) | CER: {cer:.4f} ({cer * 100:.1f}%)\n")
    print("-" * 80)

# ==========================================
# 5. Zip and Download the Audio Folder
# ==========================================
zip_name = f"{output_dir}.zip"
print(f"\nZipping '{output_dir}' and downloading to local machine...")
shutil.make_archive(output_dir, 'zip', output_dir)
files.download(zip_name)
```
