```python
from datasets import load_dataset, Audio
import soundfile as sf
import io

# 1. Load the dataset in streaming mode
dataset = load_dataset("ai4bharat/indicvoices_r", "Hindi", split="train", streaming=True)

# 2. Prevent automatic decoding of the audio column to drastically speed up the search
dataset = dataset.cast_column("audio", Audio(decode=False))

# State trackers
found_male = False
found_female = False

print("Searching stream for audio samples without pre-decoding...")

# Single-pass loop
for example in dataset:
    gender = example.get('gender')

    # Save Male
    if gender == "male" and not found_male:
        # Fetch the raw bytes
        audio_bytes = example['audio']['bytes']

        # Decode ONLY this specific file into an array using io.BytesIO
        data, samplerate = sf.read(io.BytesIO(audio_bytes))

        # Save as a clean .wav file for XTTS
        sf.write("reference_male.wav", data, samplerate)
        print(f"Saved reference_male.wav (Duration: {len(data)/samplerate:.2f}s)")
        found_male = True

    # Save Female
    elif gender == "female" and not found_female:
        # Fetch the raw bytes
        audio_bytes = example['audio']['bytes']

        # Decode ONLY this specific file
        data, samplerate = sf.read(io.BytesIO(audio_bytes))

        # Save as a clean .wav file for XTTS
        sf.write("reference_female.wav", data, samplerate)
        print(f"Saved reference_female.wav (Duration: {len(data)/samplerate:.2f}s)")
        found_female = True

    # Break early once both are found to save time
    if found_male and found_female:
        print("Both references acquired. Stopping stream.")
        break
```

```python
import os
import torch
import json

# ==========================================
# THE ESSENTIAL PATCH
# ==========================================
# We must inject this specific function back into memory
# because the newest transformers library removed it.
import transformers.pytorch_utils
if not hasattr(transformers.pytorch_utils, 'isin_mps_friendly'):
    transformers.pytorch_utils.isin_mps_friendly = torch.isin
# ==========================================

from TTS.api import TTS

# 1. Automatically agree to Coqui TOS
os.environ["COQUI_TOS_AGREED"] = "1"

print("Loading evaluation dataset...")
with open("hindi_evaluation_set.json", "r", encoding="utf-8") as f:
    eval_data = json.load(f)

print("Initializing XTTS v2 Model (this may take a minute)...")
# 2. Setup device and model
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to(device)

# 3. Prepare output directory
output_dir = "xtts_hindi_eval"
os.makedirs(output_dir, exist_ok=True)

reference_files = {
    "Male": "reference_male.wav",
    "Female": "reference_female.wav"
}

print("\nStarting Audio Generation...")

# 4. Generation Loop
for gender, ref_audio in reference_files.items():
    if not os.path.exists(ref_audio):
        print(f"Error: {ref_audio} not found. Please ensure the extraction cell ran successfully.")
        continue

    print(f"\n--- Processing {gender} Voice ---")
    for key, item in eval_data.items():
        text = item["text"]
        filename = f"{output_dir}/{item['id']}_{gender}.wav"

        print(f"Synthesizing {item['id']} ({gender})...")

        # Run inference
        tts.tts_to_file(
            text=text,
            speaker_wav=ref_audio,
            language="hi",
            file_path=filename
        )

print(f"\n✅ All done! Evaluation audio files have been saved to the '{output_dir}' folder.")
```

```python
import whisper
import jiwer
import json
import os
import pandas as pd
import warnings

# Suppress fp16 warnings from Whisper if running on CPU/older GPUs
warnings.filterwarnings("ignore")

print("Loading evaluation dataset...")
with open("hindi_evaluation_set.json", "r", encoding="utf-8") as f:
    eval_data = json.load(f)

print("Loading Whisper 'medium' model (this may take a minute to download)...")
# The 'medium' model is highly accurate for Hindi while fitting in most Colab/local VRAMs.
model = whisper.load_model("medium")

output_dir = "xtts_hindi_eval"
results = []

print("\nStarting Transcription and Evaluation...\n")

# Evaluation Loop
for gender in ["Male", "Female"]:
    for key, item in eval_data.items():
        file_id = item["id"]
        ground_truth = item["text"]
        audio_path = f"{output_dir}/{file_id}_{gender}.wav"

        if not os.path.exists(audio_path):
            print(f"Skipping {audio_path}: File not found.")
            continue

        # Transcribe using Whisper, explicitly setting the language to Hindi
        transcription_result = model.transcribe(audio_path, language="hi")
        hypothesis = transcription_result["text"].strip()

        # Calculate Error Rates (Handling potential empty transcriptions)
        try:
            wer = jiwer.wer(ground_truth, hypothesis)
            cer = jiwer.cer(ground_truth, hypothesis)
        except ValueError:
            # Fallback if Whisper outputs nothing
            wer = 1.0
            cer = 1.0

        # Store data
        results.append({
            "Test_ID": file_id,
            "Category": key,
            "Gender": gender,
            "WER": round(wer, 3),
            "CER": round(cer, 3),
            "Original_Text": ground_truth,
            "Whisper_Transcription": hypothesis
        })

        print(f"[{file_id} - {gender}] WER: {wer:.3f} | CER: {cer:.3f}")

# Create a DataFrame for Analysis
df_results = pd.DataFrame(results)

print("\n" + "="*40)
print("🏆 EVALUATION SUMMARY (Lower is Better)")
print("="*40)

# Print average scores grouped by gender
summary = df_results.groupby("Gender")[["WER", "CER"]].mean()
print(summary)

# Save detailed report to CSV
csv_filename = "xtts_whisper_evaluation_report.csv"
df_results.to_csv(csv_filename, index=False)
print(f"\n✅ Detailed row-by-row breakdown saved to '{csv_filename}'.")
```
