```python
import os, sys, json, csv, time, shutil
import torch
import soundfile as sf
from datetime import datetime

sys.path.insert(0, "/content/Spark-TTS")
from cli.SparkTTS import SparkTTS

MODEL_DIR      = "/content/Spark-TTS/pretrained_models/Spark-TTS-0.5B"
EVAL_JSON_PATH = "/content/Spark-TTS/hindi_evaluation_set.json"
OUTPUT_ROOT    = "/content/spark_tts_outputs"
SAMPLE_RATE    = 16000

DEVICE = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(f"Using device: {DEVICE}")

model = SparkTTS(MODEL_DIR, DEVICE)
print("✅ SparkTTS model loaded")

SPEAKERS = {
    "male": {
        "id":       "S4257210000350170",
        "ref_wav":  "/content/reference_audio/male_ref.wav",
        "ref_text": refs["male"]["text"],
    },
    "female": {
        "id":       "S4259138800367594",
        "ref_wav":  "/content/reference_audio/female_ref.wav",
        "ref_text": refs["female"]["text"],
    },
}

with open(EVAL_JSON_PATH, "r", encoding="utf-8") as f:
    eval_data = json.load(f)

sentences = [
    {"category": cat, "id": meta["id"], "text": meta["text"]}
    for cat, meta in eval_data.items()
]
print(f"✅ {len(sentences)} sentences loaded")

for gender in ["male", "female"]:
    os.makedirs(f"{OUTPUT_ROOT}/generated/{gender}",    exist_ok=True)
    os.makedirs(f"{OUTPUT_ROOT}/ground_truth/{gender}", exist_ok=True)

for gender in ["male", "female"]:
    for label, src in [
        ("ref", f"/content/reference_audio/{gender}_ref.wav"),
        ("gt",  f"/content/ground_truth_audio/{gender}_gt.wav"),
    ]:
        dst = f"{OUTPUT_ROOT}/ground_truth/{gender}/{gender}_{label}.wav"
        shutil.copy2(src, dst)
        print(f"  GT copied → {dst}")

results = []

for gender, spk in SPEAKERS.items():
    print(f"\n{'='*60}")
    print(f"  {gender.upper()}  |  ID: {spk['id']}")
    print(f"  Ref WAV : {spk['ref_wav']}")
    print(f"  Ref text: {spk['ref_text'][:70]}...")
    print(f"{'='*60}")

    for sent in sentences:
        out_path  = f"{OUTPUT_ROOT}/generated/{gender}/{sent['id']}_{gender}.wav"
        status    = "success"
        duration  = 0.0
        error_msg = ""

        print(f"  [{sent['id']}] {sent['text'][:52]}...", end=" ", flush=True)
        t0 = time.time()

        try:
            with torch.no_grad():
                wav = model.inference(
                    sent["text"],               # target text
                    spk["ref_wav"],             # prompt_speech_path
                    prompt_text=spk["ref_text"] # prompt transcript
                )

            sf.write(out_path, wav, samplerate=SAMPLE_RATE)
            duration = len(wav) / SAMPLE_RATE
            print(f"✅ {duration:.2f}s")

        except Exception as e:
            status    = "failed"
            error_msg = str(e)
            print(f"❌ {error_msg[:90]}")

        results.append({
            "sentence_id":          sent["id"],
            "category":             sent["category"],
            "speaker_gender":       gender,
            "speaker_id":           spk["id"],
            "text":                 sent["text"],
            "reference_wav":        spk["ref_wav"],
            "generated_audio_path": out_path if status == "success" else "",
            "gt_ref_path":          f"{OUTPUT_ROOT}/ground_truth/{gender}/{gender}_ref.wav",
            "gt_sample_path":       f"{OUTPUT_ROOT}/ground_truth/{gender}/{gender}_gt.wav",
            "duration_s":           round(duration, 3),
            "inference_time_s":     round(time.time() - t0, 3),
            "status":               status,
            "error":                error_msg,
            "timestamp":            datetime.now().isoformat(),
        })

json_path = f"{OUTPUT_ROOT}/results.json"
with open(json_path, "w", encoding="utf-8") as f:
    json.dump(results, f, ensure_ascii=False, indent=2)
print(f"\n📄 JSON → {json_path}")

csv_path = f"{OUTPUT_ROOT}/results.csv"
with open(csv_path, "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=list(results[0].keys()))
    writer.writeheader()
    writer.writerows(results)
print(f"📊 CSV  → {csv_path}")

ok   = sum(1 for r in results if r["status"] == "success")
fail = len(results) - ok
print(f"\n{'─'*62}")
print(f"{'ID':<8} {'Gender':<8} {'Dur':>7}  {'Inf.Time':>9}  Status")
print(f"{'─'*62}")
for r in results:
    flag = "✅" if r["status"] == "success" else "❌"
    print(f"{r['sentence_id']:<8} {r['speaker_gender']:<8} "
          f"{r['duration_s']:>6.2f}s  {r['inference_time_s']:>8.2f}s  {flag}")
print(f"{'─'*62}")
print(f"  {ok} success  |  {fail} failed  |  {len(results)} total")

!zip -r /content/spark_tts_outputs.zip /content/spark_tts_outputs
from google.colab import files
files.download("/content/spark_tts_outputs.zip")
```
