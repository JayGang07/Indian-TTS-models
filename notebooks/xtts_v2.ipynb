```
import torch
import os
from TTS.api import TTS

# Bypass the Terms of Service prompt
os.environ["COQUI_TOS_AGREED"] = "1"

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}\n")

print("Loading XTTS v2 Model (This may take a minute to download)...")
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to(device)

text_to_speak = "हैलो, मेरा नाम जय है।"
reference_audio_path = "Demo-Audio2.wav"  # Replace with your actual reference audio file

print("Cloning voice and generating speech...")
tts.tts_to_file(
    text=text_to_speak,
    speaker_wav=reference_audio_path,
    language="hi",
    file_path="xtts_output.wav"
)

print("✅ Voice cloned successfully! Saved to xtts_output.wav")
```
