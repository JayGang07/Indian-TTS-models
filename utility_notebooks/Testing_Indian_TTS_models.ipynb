```python
import torch
from huggingface_hub import login
from transformers import AutoTokenizer, VitsModel
from parler_tts import ParlerTTSForConditionalGeneration
import soundfile as sf
import scipy.io.wavfile as wavfile

# --- 1. Authenticate ---
# Replace with your actual token
my_token = "YOUR_HF_TOKEN_HERE"
login(token=my_token)

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"\nUsing device: {device}\n")

# ==========================================
# --- 2. TEST INDIC PARLER-TTS ---
# ==========================================
print("Loading Indic Parler-TTS model...")
parler_id = "ai4bharat/indic-parler-tts"

# 🚨 CRITICAL FIX: Load TWO separate tokenizers
text_tokenizer = AutoTokenizer.from_pretrained(parler_id, use_fast=False) # For the Indic text
desc_tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-large")    # For the English description

parler_model = ParlerTTSForConditionalGeneration.from_pretrained(parler_id).to(device)

# The Inputs
text_to_speak = "हैलो, मेरा नाम जय है।"
voice_description = "A male speaker with a clear, natural voice talking at a calm, steady pace in a quiet room."

# 🚨 CRITICAL FIX: Map them correctly!
# Description goes to input_ids (via desc_tokenizer)
desc_tokens = desc_tokenizer(voice_description, return_tensors="pt")
input_ids = desc_tokens.input_ids.to(device)
input_mask = desc_tokens.attention_mask.to(device)

# Text to speak goes to prompt_input_ids (via text_tokenizer)
text_tokens = text_tokenizer(text_to_speak, return_tensors="pt")
prompt_ids = text_tokens.input_ids.to(device)
prompt_mask = text_tokens.attention_mask.to(device)

print("Generating Parler-TTS audio...")
with torch.no_grad():
    generation = parler_model.generate(
        input_ids=input_ids,
        attention_mask=input_mask,
        prompt_input_ids=prompt_ids,
        prompt_attention_mask=prompt_mask,
    )

audio_arr_p = generation.cpu().numpy().squeeze()
sf.write("indic_parler_output.wav", audio_arr_p, parler_model.config.sampling_rate)
print("✅ Saved Parler-TTS to indic_parler_output.wav\n")

```

```python
# ==========================================
# --- 3. TEST VITS RASA 13 ---
# ==========================================
from transformers import AutoModel # Note: AutoModel, not VitsModel

print("Loading VITS Rasa 13 model...")
# Fixed the underscores
vits_id = "ai4bharat/vits_rasa_13"

# Added trust_remote_code=True to both
vits_tokenizer = AutoTokenizer.from_pretrained(vits_id, trust_remote_code=True)
vits_model = AutoModel.from_pretrained(vits_id, trust_remote_code=True).to(device)

text_v = "हैलो, मेरा नाम जय है।"
inputs_v = vits_tokenizer(text_v, return_tensors="pt").to(device)

print("Generating VITS Rasa 13 audio...")
with torch.no_grad():
    # VITS Rasa 13 accepts both speaker_id and emotion_id
    speaker_id = 0  # 0 for ASM_F, 1 for ASM_M, etc.
    style_id = 0    # 0 for ALEXA, 1 for ANGER, etc.

    output = vits_model(
        inputs_v['input_ids'],
        speaker_id=speaker_id,
        emotion_id=style_id
    )

audio_arr_v = output.waveform.cpu().numpy().squeeze()
wavfile.write("vits_rasa_output.wav", vits_model.config.sampling_rate, audio_arr_v)
print("✅ Saved VITS to vits_rasa_output.wav")
```

```python
import torch
import os
from huggingface_hub import login
from transformers import AutoTokenizer, AutoModel
from parler_tts import ParlerTTSForConditionalGeneration
import soundfile as sf
import scipy.io.wavfile as wavfile

# 1. 🚨 RE-AUTHENTICATE HERE (Keep the quotation marks!)
my_token = "YOUR_HF_TOKEN_HERE"
print("Logging into Hugging Face...")
login(token=my_token)

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}\n")

# ==========================================
# 2. RUN PARLER-TTS TEST
# ==========================================
print("Loading Indic Parler-TTS model...")
parler_id = "ai4bharat/indic-parler-tts"

text_tokenizer = AutoTokenizer.from_pretrained(parler_id, use_fast=False)
desc_tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-large")
parler_model = ParlerTTSForConditionalGeneration.from_pretrained(parler_id).to(device)

text_to_speak = "नमस्ते, यह भारतीय भाषा तकनीक के परीक्षण का एक उदाहरण है।"
voice_description = "A female speaker with a clear, natural voice talking at a calm, steady pace in a quiet room."

desc_tokens = desc_tokenizer(voice_description, return_tensors="pt")
text_tokens = text_tokenizer(text_to_speak, return_tensors="pt")

print("Generating Parler-TTS audio...")
with torch.no_grad():
    generation = parler_model.generate(
        input_ids=desc_tokens.input_ids.to(device),
        attention_mask=desc_tokens.attention_mask.to(device),
        prompt_input_ids=text_tokens.input_ids.to(device),
        prompt_attention_mask=text_tokens.attention_mask.to(device),
    )

audio_arr_p = generation.cpu().numpy().squeeze()
sf.write("indic_parler_output.wav", audio_arr_p, parler_model.config.sampling_rate)
print("✅ Saved Parler-TTS to indic_parler_output.wav\n")

# ==========================================
# 3. RUN VITS RASA 13 TEST
# ==========================================
print("Loading VITS Rasa 13 model...")
vits_id = "ai4bharat/vits_rasa_13"

vits_tokenizer = AutoTokenizer.from_pretrained(vits_id, trust_remote_code=True)
vits_model = AutoModel.from_pretrained(vits_id, trust_remote_code=True).to(device)

inputs_v = vits_tokenizer(text_to_speak, return_tensors="pt").to(device)

print("Generating VITS Rasa 13 audio...")
with torch.no_grad():
    output = vits_model(
        inputs_v['input_ids'],
        speaker_id=1,
        emotion_id=0
    )

audio_arr_v = output.waveform.cpu().numpy().squeeze()
wavfile.write("vits_rasa_output.wav", vits_model.config.sampling_rate, audio_arr_v)
print("✅ Saved VITS to vits_rasa_output.wav")
```
